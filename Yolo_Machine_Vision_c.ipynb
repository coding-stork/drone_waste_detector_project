```python
import os, random, math, shutil
!mkdir val test
content = os.getcwd()
imgdir = os.listdir(os.path.join(content,"train"))
split_var = math.floor(len(imgdir)*0.1)
for i in range(split_var):
  elem = random.choice(imgdir)[:-4]
  os.replace(os.path.join(content,"train",elem+".jpg"),os.path.join(content,"val",elem+".jpg"))
  os.replace(os.path.join(content,"train",elem+".txt"),os.path.join(content,"val",elem+".txt"))
  imgdir.remove(elem+".jpg")
  imgdir.remove(elem+".txt")
for i in range(split_var):
  elem = random.choice(imgdir)[:-4]
  os.replace(os.path.join(content,"train",elem+".jpg"),os.path.join(content,"test",elem+".jpg"))
  os.replace(os.path.join(content,"train",elem+".txt"),os.path.join(content,"test",elem+".txt"))
  imgdir.remove(elem+".jpg")
  imgdir.remove(elem+".txt")
```
